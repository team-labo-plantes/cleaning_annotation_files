```python
# required librairies
import pandas as pd
import io 
```

```python
# define the input file 
input_timar_path = r'C:\Users\Lecab\gitrepo\plantes_ornementales\ornamental\clematis_dbgi\pos\results\tima-r\230922_155606_dbgi_clematis_230922\annotations.tsv'
```

```python
# formatting the datafile 

df_timar = pd.read_csv(input_timar_path, sep='\t') 
suffix = "_timar"

# Adding _timar suffix to all columns
df_timar.columns = [col + suffix for col in df_timar.columns]

df_timar.head(10)
```

```python
# Filter according the final score 
# Create a new column "first_score_final_timar' extractting the first value 
df_timar['first_score_final_timar'] = df_timar['score_final_timar'].str.split('|').str[0]
```

```python
# List of columns to keep
columns_to_keep = ['feature_id_timar', 'feature_mz_timar', 'feature_rt_timar', 'candidate_structure_molecular_formula_timar',	'candidate_library_timar', 'candidate_count_similarity_peaks_matched_timar', 'candidate_score_similarity_timar', 'candidate_structure_exact_mass_timar', 'candidate_structure_xlogp_timar', 'candidate_structure_inchikey_no_stereo_timar', 'candidate_structure_smiles_no_stereo_timar', 'candidate_structure_tax_cla_chemontid_timar', 'candidate_structure_tax_cla_01kin	candidate_structure_tax_cla_02sup_timar',	'candidate_structure_tax_cla_03cla_timar',	'candidate_structure_tax_cla_04dirpar_timar',	'candidate_structure_tax_npc_01pat_timar', 'candidate_structure_tax_npc_02sup_timar', 'candidate_structure_tax_npc_03cla_timar',	'candidate_structure_organism_occurrence_closest_timar',	'candidate_structure_organism_occurrence_reference_timar', 'candidate_structure_error_mz_timar', 'candidate_structure_error_rt_timar', 'rank_initial_timar', 'rank_final_timar', 'score_initial_timar', 'score_biological_timar', 'score_chemical_timar',	'score_final_timar', 'first_score_final_timar']

# Drop unwanted columns
columns_to_drop = [col for col in df_timar.columns if col not in columns_to_keep]
df_timar = df_timar.drop(columns_to_drop, axis=1)

print('TimaR file is cleaned !')

#save the new dataframe as a tsv file under the name canopus_formula_summary_cleaned.tsv
output_timar_path = r'C:\Users\Lecab\gitrepo\plantes_ornementales\ornamental\clematis_dbgi\pos\results\tima-r\230922_155606_dbgi_clematis_230922\annotations_cleaned.tsv'
df_timar.to_csv(output_timar_path, sep='\t')

df_timar.head(10)
```
